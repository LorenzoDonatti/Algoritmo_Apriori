```python
import pandas as pd
import numpy as np

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df1 = pd.read_csv("http://www-usr.inf.ufsm.br/~joaquim/UFSM/DM/ds/_ASSOC_BGFriends_01.csv", encoding='utf8', sep=',')
```

```python
df2 = pd.read_csv("http://www-usr.inf.ufsm.br/~joaquim/UFSM/DM/ds/_ASSOC_BGFriends_02.csv", encoding='latin1', sep=',')
```

```python
display(df1)
```

```python
display(df2)
```

```python
df = pd.concat([df1,df2])
display(df)
```

```python
df.drop(columns=['Partida','Oponentes','Amigos'],inplace=True)
display(df)
```

```python
print(df['Jogadore(a)s'].value_counts())
print(df['Jogadore(a)s'].unique())
```

```python
df['Jogadore(a)s'] = df['Jogadore(a)s'].str.replace("ç", "c")
df['Jogadore(a)s'] = df['Jogadore(a)s'].str.replace(" ", "")
df['Jogadore(a)s'] = df['Jogadore(a)s'].str.upper()

print(df['Jogadore(a)s'].value_counts())
print(df['Jogadore(a)s'].unique())
```

```python
lista= df['Jogadore(a)s'].tolist()
print(lista)

for i in range(len(lista)):
  lista[i] = lista[i].split(',')

lista
```

```python
te = TransactionEncoder()
te_ary = te.fit(lista).transform(lista)
dft = pd.DataFrame(te_ary, columns=te.columns_)
dft
```

```python
frequent_itens = apriori(dft, min_support = 0.1, use_colnames=True)
frequent_itens
```
